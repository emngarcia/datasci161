The following treatment matrix (using data from [here](https://sites.socsci.uci.edu/~dneumark/datasets.html)) has states as rows and year-months as columns, with 1s indicating treated periods (when state MW > federal minimum). The federal minimum wage last increased in 2009, so we will look at timepoints from 2010 onwards.

```python
import pandas as pd
import numpy as np

# Read the minimum wage data
mw_data = pd.read_csv('/Users/ElizabethMiriya/Downloads/MW_LW dataset updated through 2019 - 1-01-20 Update (1).csv')

# Convert Year, Quarter, Month to datetime for easier handling
mw_data['date'] = pd.to_datetime(mw_data['Year'].astype(str) + '-' +
                                  mw_data['Month'].astype(str),
                                  format='%Y-%m', errors='coerce')

# Create a year-month identifier
mw_data['year_month'] = mw_data['Year'].astype(str) + '-' + mw_data['Month'].astype(str).str.zfill(2)

# For treatment definition, we need to know the federal minimum wage
# Federal minimum has been $7.25 since July 2009
FEDERAL_MW = 7.25

# Identify if state MW is above federal (treated)
mw_data['above_federal'] = (mw_data['MW'] > FEDERAL_MW).astype(int)

# Filter to 2010 onwards
mw_post2010 = mw_data[mw_data['Year'] >= 2010].copy()

#change column name

mw_post2010 = mw_post2010.rename(columns={'State Name': 'state'})

# Create treatment indicator: state is treated starting from first month it raised MW above federal
treatment_status = []

for state in mw_post2010['state'].unique():
    state_data = mw_post2010[mw_post2010['state'] == state].sort_values('Year')

    # Find first instance where MW > federal after 2010
    first_treatment = state_data[state_data['above_federal'] == 1]

    if len(first_treatment) > 0:
        treatment_year = first_treatment.iloc[0]['Year']
        treatment_month = first_treatment.iloc[0]['Month']

        # Mark as treated from that point forward
        state_data['treated'] = 0
        state_data.loc[(state_data['Year'] > treatment_year) |
                       ((state_data['Year'] == treatment_year) &
                        (state_data['Month'] >= treatment_month)), 'treated'] = 1
    else:
        state_data['treated'] = 0

    treatment_status.append(state_data)

treatment_long = pd.concat(treatment_status, ignore_index=True)

# Create treatment matrix: rows = states, columns = year-months, values = treatment status
dfw = treatment_long.pivot_table(
    index='state',
    columns='year_month',
    values='treated',
    aggfunc='first'  # Take first value if duplicates
)

# Sort columns chronologically
dfw = dfw[sorted(dfw.columns)]

# Fill any missing values with 0 (not treated)
dfw = dfw.fillna(0).astype(int)

dfw

# Save to CSV
#dfw.to_csv('mw_treatment_panel.csv')
```

```python
# making dfw fit dfy

# Step 1: Extract only January columns
january_cols = [col for col in dfw.columns if col.endswith('-01')]
df_january = dfw[january_cols].copy()

# Step 2: Pick specific years manually
selected_years = years  # years with outcome data (between 2010 and 2019)
selected_cols = [col for col in df_january.columns if col[:4] in selected_years]

# Step 3: Subset the matrix
w_matrix = df_january[selected_cols].copy()

# Step 4: Rename columns to just the year (remove '-01')
w_matrix = w_matrix.rename(columns={col: col[:4] for col in w_matrix.columns})
w_matrix.columns.name = "year"

w_matrix
```

```python
# Keep only rows that exist in both matrices
common_rows = y_matrix.index.intersection(w_matrix.index)
w_matrix = w_matrix.loc[common_rows]
y_matrix = y_matrix.loc[common_rows]
w_matrix.shape
w_matrix
```

```python
# Identify rows that are fully treated
rows_all_treated = np.all(w_matrix == 1, axis=1)

# Keep only rows that are not fully treated
y_matrix = y_matrix[~rows_all_treated.values]
w_matrix = w_matrix[~rows_all_treated.values]
w_matrix
```

# Difference-in-Differences

```python
# !pip install linearmodels
from linearmodels.panel import PanelOLS
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# 1. Convert matrices to long format
# ------------------------------

dfy_long = (
    y_matrix
    .reset_index()
    .melt(id_vars=y_matrix.index.name, var_name='year', value_name='Y')
)

dfw_long = (
    w_matrix
    .reset_index()
    .melt(id_vars=w_matrix.index.name, var_name='year', value_name='D')
)

dfy_long['year'] = dfy_long['year'].astype(int)
dfw_long['year'] = dfw_long['year'].astype(int)

# ------------------------------
# 2. Merge outcome + treatment
# ------------------------------

df = dfy_long.merge(dfw_long, on=[y_matrix.index.name, 'year'])

# ------------------------------
# 3. Compute first treated year
# ------------------------------

first_treat = (
    df.query("D == 1")
    .groupby(y_matrix.index.name)['year']
    .min()
    .rename("first_treat_year")
)
df = df.merge(first_treat, on=y_matrix.index.name, how='left')

# ------------------------------
# 4. Compute event time
# ------------------------------

df['event_time'] = df['year'] - df['first_treat_year']

# ------------------------------
# 5. Build event-time dummies (omit k=-1)
# ------------------------------

unique_event_times = df['event_time'].dropna().unique()
for k in unique_event_times:
    if k == -1:
        continue  # baseline
    df[f'event_{int(k)}'] = (df['event_time'] == k).astype(int)

# ------------------------------
# 6. Set panel index
# ------------------------------

df_panel = df.set_index([y_matrix.index.name, 'year'])
# Replace minus sign with 'm', so event_-2 -> event_m2
df_panel = df_panel.rename(columns=lambda x: x.replace('event_-', 'event_m') if 'event_-' in x else x)


# export df to use for Callaway & Sant'Anna DiD Analysis in R

df_panel.to_csv("/Users/ElizabethMiriya/Desktop/did_df_crime.csv", index=True)

df_panel
```

# Matrix completion

method is from lightweight-mcnnm, cites athey et al (2021)

```python
import numpy as np
import jax.numpy as jnp
from mcnnm import estimate   # lightweight-mcnnm's main wrapper

# 1. Make sure indices & columns line up across Y and W
# (they should already, but this keeps things safe)
y_aligned = y_matrix.sort_index().sort_index(axis=1)
w_aligned = w_matrix.loc[y_aligned.index, y_aligned.columns]

# 2. Convert to JAX arrays
Y = jnp.array(y_aligned.to_numpy(), dtype=jnp.float64)
W = jnp.array(w_aligned.to_numpy(), dtype=jnp.int32)

print("Y shape:", Y.shape)   # (N_states, T_years)
print("W shape:", W.shape)
```

```python
# Run MC-NNM with cross-validated penalties
results = estimate(
    Y=Y,
    Mask=W,                # 1 = treated, 0 = untreated
    Omega=None,            # all outcomes observed
    use_unit_fe=True,      # include unit fixed effects
    use_time_fe=True,      # include time fixed effects
    lambda_L=None,         # let CV pick low-rank penalty
    lambda_H=None,         # let CV pick high-rank penalty for high-rank noise
    validation_method='cv',
    K=5,                   # 5-fold CV
    n_lambda=20,           # number of candidate lambdas
)

# Main outputs
tau_mc = float(results.tau)                # overall ATT
Y_completed_jax = results.Y_completed      # completed outcome matrix (JAX array)

print(f"MC-NNM estimated ATT (tau): {tau_mc:.3f}")
print(f"Chosen lambda_L: {float(results.lambda_L):.4f}")
print(f"Chosen lambda_H: {float(results.lambda_H):.4f}")
```

```python
import numpy as np

W_np = np.array(w_aligned.to_numpy(), dtype=int)
state_names = np.array(y_aligned.index)

# list of treated units
treated_units = np.where(W_np.sum(axis=1) > 0)[0]  # any row with some 1's
control_units = np.where(W_np.sum(axis=1) == 0)[0]

# store each unit's treatment timing pattern
treat_patterns = {i: W_np[i, :].copy() for i in treated_units}

tau_obs = tau_mc

tau_placebos = []

for t_unit in treated_units:
    pattern = treat_patterns[t_unit]   # timing pattern of this treated unit

    for c in control_units:
        # placebo W matrix
        W_pl = np.zeros_like(W_np)
        W_pl[c, :] = pattern           # assign treatment pattern to control unit

        W_pl_jnp = jnp.array(W_pl, dtype=jnp.int32)

        # run matrix completion
        res_pl = estimate(
            Y=Y,
            Mask=W_pl_jnp,
            Omega=None,
            use_unit_fe=True,
            use_time_fe=True,
            lambda_L=None,
            lambda_H=None,
            validation_method='cv',
            K=5,
            n_lambda=20,
        )

        tau_placebos.append(float(res_pl.tau))

tau_placebos = np.array(tau_placebos)
print("Placebo taus:", tau_placebos)

p_val = np.mean(np.abs(tau_placebos) >= abs(tau_obs))

print(f"Observed tau: {tau_obs:.3f}")
print(f"Permutation p-value: {p_val:.3f}")
print(f"Mean placebo tau: {tau_placebos.mean():.3f}")
print(f"SD placebo tau: {tau_placebos.std():.3f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- 0. Convert JAX arrays to NumPy
Y_np = np.array(Y)                       # observed outcomes
W_np = np.array(W)                       # 1 = treated, 0 = untreated
Y_cf = np.array(results.Y_completed)     # MC-NNM counterfactual (no-treatment) matrix

states = y_matrix.index.tolist()         # state names
years = y_matrix.columns.astype(int).tolist()

# --- 1. tau_it: cell-level treatment effects
# tau_it[i,t] = (Y - counterfactual) for treated cells, NaN otherwise
tau_it = (Y_np - Y_cf)
tau_it[W_np == 0] = np.nan

unit_labels = states
time_labels = [str(y) for y in years]

tau_it_df = pd.DataFrame(tau_it, index=unit_labels, columns=time_labels)

# --- 2. Unit ATT per unit (average over treated periods for that unit)
unit_att = np.full(Y_np.shape[0], np.nan)
for i in range(Y_np.shape[0]):
    treated_mask = W_np[i, :] == 1
    if treated_mask.any():
        unit_att[i] = np.nanmean(tau_it[i, treated_mask])

unit_att_df = pd.DataFrame({
    "Unit": states,
    "Unit_ATT": unit_att
})

print("Average post-treatment effect per unit:")
print(unit_att_df)

print("\nUnit  time treatment effects (tau_it):")
print(tau_it_df)

# --- 3. Event-study: average effect by time relative to first treatment

# Put W and tau_it in long format
w_df = pd.DataFrame(W_np, index=states, columns=years)
tau_df = pd.DataFrame(tau_it, index=states, columns=years)

df_long = (
    w_df.stack().rename("D").to_frame()
        .join(tau_df.stack().rename("tau_it"))
        .reset_index()
        .rename(columns={"level_0": "state", "level_1": "year"})
)

df_long["year"] = df_long["year"].astype(int)

# First treated year per state
first_treat = (
    df_long.query("D == 1")
           .groupby("state")["year"]
           .min()
           .rename("first_treat_year")
)

df_long = df_long.merge(first_treat, on="state", how="left")

# Event time (relative to first treatment)
df_long["event_time"] = df_long["year"] - df_long["first_treat_year"]

# Keep treated periods with valid tau_it
df_treated = df_long[(df_long["D"] == 1) & df_long["tau_it"].notna()]

# Average tau_it by event_time
event_avg_series = df_treated.groupby("event_time")["tau_it"].mean()

# Build a complete relative-time index (for gaps)
min_k = int(event_avg_series.index.min())
max_k = int(event_avg_series.index.max())
relative_time = np.arange(min_k, max_k + 1)

event_avg = event_avg_series.reindex(relative_time).to_numpy()

max_lead = -min_k
max_lag = max_k

event_avg_df = pd.DataFrame({
    "Relative_Time": relative_time,
    "Event_Avg": event_avg
})

print("\nEvent-study average by relative period:")
print(event_avg_df)
```
